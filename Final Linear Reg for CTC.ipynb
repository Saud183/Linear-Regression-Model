```python
#installing dependencies
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_train = pd.read_csv(Train Data Path)
data_test = pd.read_csv(Test Data Path)
```

```python
data_train
```

```python
data_test
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
train_par = data_train.drop(['S.No.','CTC','College','City type','Role'], axis=1)
train_param = train_par.apply(pd.to_numeric, errors='coerce')

train_param
```

```python
train_c = data_train['CTC']
train_ctc = train_c.apply(pd.to_numeric, errors='coerce')

train_ctc
```

```python
x_train, x_test, y_train, y_test = train_test_split(train_param, train_ctc)
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
```

```python
data_test
```

```python
test_da = data_test.drop(['College','Role','City type','College_T1','College_T2','Role_Manager','City_Metro','Actual CTC','Predicted CTC'], axis=1)

test_data = test_da.apply(pd.to_numeric, errors='coerce')

test_data
```

```python
pred = regr.predict(test_data)
```

```python
pred
```

```python
data_test.loc[:,'Predicted CTC'] = pred
```
